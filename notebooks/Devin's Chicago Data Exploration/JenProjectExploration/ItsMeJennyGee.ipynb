```python
import pandas as pd
import numpy as np
```

```python
df_crimes = pd.read_csv("Crimes_-_One_year_prior_to_present.csv")
```

```python
df_crimes
```

```python
df_crimes.columns
```

```python
df_crimes.columns = [x.lstrip().replace(' ', "_").lower() for x in df_crimes.columns]
```

```python
df_crimes.columns
```

```python
df_crimes['date__of_occurrence'] = pd.to_datetime(df_crimes['date__of_occurrence'])
```

```python
df_crimes = df_crimes.sort_values(by = 'date__of_occurrence').reset_index()
```

```python
df_crimes['primary_description'].unique()
```

```python
df_crimes.groupby('primary_description').count().iloc[:,0]
```

```python
df_violent_crimes = df_crimes[~df_crimes['primary_description'].isin(['DECEPTIVE PRACTICE','OTHER OFFENSE','WEAPONS VIOLATION','CRIMINAL TRESPASS',
                                                                      'NARCOTICS','PUBLIC PEACE VIOLATION', 'PROSTITUTION',
                                                                      'INTERFERENCE WITH PUBLIC OFFICER', 'INTIMIDATION','CONCEALED CARRY LICENSE VIOLATION',
                                                                      'LIQUOR LAW VIOLATION', 'OTHER NARCOTIC VIOLATION', 'OBSCENITY','PUBLIC INDECENCY', 'NON-CRIMINAL',
                                                                      'GAMBLING', 'RITUALISM'])]
```

```python
df_max_temps = pd.read_csv("ChicagoMaxTemps.csv")[['Date',"Maximum Temperature"]].dropna()
```

```python
df_violent_crimes['Date'] = df_violent_crimes.date__of_occurrence.dt.date
```

```python
df_max_temps["Date"] = pd.to_datetime(df_max_temps["Date"])
```

```python
df_violent_crimes['Date'] = df_violent_crimes.Date.astype('datetime64[ns]')
df_max_temps['Date'] = df_max_temps.Date.astype('datetime64[ns]')
```

```python
df_crimes_temp = df_violent_crimes.merge(df_max_temps, on="Date")
```

```python
df_crimes_temp
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
```

```python
df_crimes_temp[['x_coordinate', 'y_coordinate', 'latitude', 'longitude', 'Maximum Temperature']].info()
```

```python
df_crimes_temp['homicide'] = df_crimes_temp['primary_description'] == 'HOMICIDE'
```

```python
df_crimes_temp['homicide'].value_counts()
```

```python
features = df_crimes_temp[['x_coordinate', 'y_coordinate', 'latitude', 'longitude', 'homicide']].dropna()
x = features[['x_coordinate', 'y_coordinate', 'latitude', 'longitude']]
y = features['homicide']
```

```python
rf = RandomForestClassifier(n_estimators=20)
np.mean(cross_val_score(rf, x, y, cv=5))
```

```python
features = df_crimes_temp[['x_coordinate', 'y_coordinate', 'latitude', 'longitude', 'Maximum Temperature', 'homicide']].dropna()
x = features[['x_coordinate', 'y_coordinate', 'latitude', 'longitude', 'Maximum Temperature']]
y = features['homicide']
```

```python
rf = RandomForestClassifier(n_estimators=20)
np.mean(cross_val_score(rf, x, y, cv=5))
```
